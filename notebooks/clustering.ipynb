```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import ast
```

```python
X = pd.read_csv('../data/web_embedding.csv')
X
```

```python
# Detect format and clean accordingly
def clean_embedding(embedding_str):
    if ',' in embedding_str:
        # Case A: Python list string
        return ast.literal_eval(embedding_str)
    else:
        # Case B: NumPy-style array string
        return np.fromstring(embedding_str.strip("[]"), sep=' ')

# Apply the cleaning function
X["embedding"] = X["embedding"].apply(clean_embedding)

# Convert to a 2D NumPy array
embedding_array = np.array(X["embedding"].tolist(), dtype=np.float32)

```

```python
# Set the number of clusters
k = 300

# Run KMeans clustering on the embeddings
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(embedding_array)

# Save the cluster centroids to a .npy file
np.save("centroids.npy", kmeans.cluster_centers_)

# Add the cluster label to the original dataframe
X["cluster"] = kmeans.labels_

# Save the updated dataframe with cluster labels
X.to_csv("verses_with_embeddings_and_clusters.csv", index=False)

print(f"Clustering complete. Centroids saved to 'centroids.npy'.")
print(f"Verses with cluster labels saved to 'verses_with_clusters.csv'.")
```
